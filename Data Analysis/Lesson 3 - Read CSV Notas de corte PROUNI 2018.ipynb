## Groupment Command
If you need group results, use the dataframe method `groupby(x)[y].z()`, where we can interpreter:

> Agroup one attribute x by y. However, apply the rule z (min, max, count, etc)

**Example: Agroup the attribute `uf_busca` by highest 'mensalidade`**

Solution: 
```python
data.groupby('uf_busca')['mensalidade'].max()
```
