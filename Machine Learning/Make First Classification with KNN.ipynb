# Make First Classification with KNN

## 1. Importing libraries from `matplotlip` and `sklearn`

```python
%matplotlib notebook

import matplotlib.pyplot as plt
from sklearn.datasets import make_classification # Make classification model
```

## 2. Data plot config

Parameters description of `make_classification`:

* `n_samples` points number for plotting
* `n_features` atributtes number for object
* `n_classes` classes amount of model


```python
plt.figure(figsize=(4,4))
plt.subplot(111)

# Classification config
X1, Y1 = make_classification(n_samples=20, n_features=5, n_redundant=0, n_informative=2,
                            n_clusters_per_class=1, random_state=3, class_sep=1.1, n_classes=3)

# Plot exibition
plt.scatter(X1[:,0], X1[:,1], marker="o", c=Y1, s=25, edgecolor="K")
plt.show()
```

## 3. Create a classification algoritm

For create a classification algoritm, is necessary import function for divide test and train data

* `X` input data
* `Y` output data

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

knn = KNeighborsClassifier(n_neighbors = 1)
x_train, x_test, y_train, y_test = train_test_split(X1, Y1, random_state = 1)

print("Data view:")
print(train_test_split(X1,Y1,random_state=1))

print("Train data (x_train):")
print(x_train)

print("Test data (x_test):")
print(x_test)

print("Train data (y_train):")
print(y_train)

print("Test data (y_teste):")
print(y_test)
```

Knowing of data view structure above, is necesseary show the model create with this data. Type the function `knn.fit()` for this

```python
knn.fit(x_train, y_train) # Placing the train data (X, Y) in model
```

## 4. Verifying score 

Is necessary verify score of data train and tests in created model. Its means that we have one idea about the amount of rate success in this model

```python
print("Train rate: {}".format(knn.score(x_train, y_train)))
print("Test rate: {}".format(knn.score(x_test, y_test)))
```
